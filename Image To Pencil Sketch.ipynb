# Image to pencilsketch with python

```python
#import the libraries
import os
os.sys.path
import cv2
from PIL import Image
from IPython.display import display
```

```python
#Importing and Resizing the Image
img = cv2.imread('/content/drive/MyDrive/Lets Grow More/Subhankar Samanta LinkedIn Photo.png', 1)
img = cv2.resize(img, (480, 480))
```

```python
#Plotting the Image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

```python
#Conerting the 3 channel image to 1 channel image (GreyScale)
greyscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale))
```

```python
#Inverting the Image
inv_img = cv2.bitwise_not(greyscale)
display(Image.fromarray(inv_img))
```

```python
#Applying Gausain Blur to Sofetn the Image
soft_img = cv2.GaussianBlur(inv_img, (5, 5), 0, 0)
display(Image.fromarray(soft_img))
```

```python
#Inverting the prev image
soft_inv_Img = 255 - soft_img
final_Sketch = cv2.divide(greyscale, soft_inv_Img, scale=200.0)
display(Image.fromarray(final_Sketch))
```

```python
#Comparing the Input Image and Pencil Sketch Image
display(Image.fromarray(img))
display(Image.fromarray(final_Sketch))
```
